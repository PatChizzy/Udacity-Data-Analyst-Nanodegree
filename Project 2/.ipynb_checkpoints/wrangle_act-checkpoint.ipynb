### Use tweepy to extract 
1. retweet count
2. favourite count
3. full tweet
using tweet_id from the archive twitter data of we rate dogs

------
First set up my tweepy api and config

```python
config = configparser.ConfigParser()
config.read('/content/drive/MyDrive/Colab Notebooks/ALX x Udacity Data Analyst/Project 2/config.ini')
```

```python
api_key = config['twitter']['api_key']
api_key_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']

auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

Extract favourite count and retweet count of tweets

```python
import json

# Open a file in write mode
with open("tweet_json.txt", "w") as file:
    # Loop through the tweets in the WeRateDogs Twitter archive
    for tweet_id in archive['tweet_id']:
        try:
            # Get the tweet object from the API
            tweet = api.get_status(tweet_id)

            # Get the created_at attribute of the tweet
            created_at = tweet.created_at

            # Convert the created_at attribute to a string using the strftime method
            created_at_str = created_at.strftime("%Y-%m-%d %H:%M:%S")

            # Create a dictionary with the tweet data
            tweet_data = {
                "retweet_count": tweet.retweet_count,
                "favorite_count": tweet.favorite_count,
                "created_at": created_at_str,
                "tweet_id": tweet.id
            }

            # Write the tweet data to the file using the json.dump function
            json.dump(tweet_data, file)
            file.write("\n")

        except tweepy.TweepError as error:
            # If the error code is 144, it indicates that the tweet does not exist or is not available
            if error.api_code == 144:
                print("Skipping tweet with ID", tweet_id, "because it is not available")
                continue

            elif error.api_code == 179:
              print("Skipping tweet with ID", tweet_id, "because of authorization issue")
              continue

            else:
                # If the error code is not 144 or 179, it indicates a different problem
                print("Error when trying to get tweet with ID", tweet_id, ":", error.reason)
                continue

# Close the file
file.close()
```

```python
def count_rows(df, column_name, value):
    ''' function to count values in a row'''
    
    # Select only the rows where the specified column is equal to the specified value
    selected_rows = df.loc[df[column_name] == value]

    # Get the number of rows and columns in the DataFrame
    row_count = selected_rows.shape

    # Print the number of rows
    print("There are", row_count[0], "rows where the", column_name, "column is equal to", value)
```

###### Define
Cleaning outliers

###### Code

```python
def outliers_detector(df):
    ''' function to find outliers in a dataset and replace with the mean of that column'''

    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    upper = df[~(df > (q3 + 1.5 * IQR))].max()
    lower = df[~(df < (q1 - 1.5 * IQR))].min()
    df = np.where(df > upper, df.mean(), 
                 np.where(df < lower, df.mean(), df))
    return df
```

```python
# applying the function on the rating numerator and denominator columns in the twitter archived data
archive_clean['rating_denominator'] = outliers_detector(archive_clean['rating_denominator'])
archive_clean['rating_numerator'] = outliers_detector(archive_clean['rating_numerator'])
```

The average rating numerator is 11. However, as the days go by, the ratings are seen to exceed avaerage
